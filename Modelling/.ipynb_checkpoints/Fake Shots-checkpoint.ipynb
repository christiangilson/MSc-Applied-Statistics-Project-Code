### Shot Success

**Now we can start to look at some outliers.**

* Some analyses would completely remove outliers - even though the data points really happened - because they degrade the performance of a model.
* My philosophy is to remove outliers only if those data points are indeed errornous, and the outlier-nature of the datapoint enabled the finding of an error
* If outliers really happened, then they highlight the extremes of what can happen.
* And if this causes the data to be misrepresentative, then we should include this as a prior, not by altering the truely observed data.

And so the purpose of this analysis is to identify areas of the pitch where we may wish to provide a shot/goal prior. We clearly won't need to provide this prior for shots taken on the penalty spot. **We're looking at the real outliers: one shot one goal from nowhere which will greatly impact the payoff of moving a pass into that zone**.

Why is this important?
* Will definitely impact xT;
* Will therefore possibly bias our valuation of players actions purely by the luck of whether they make a pass on a square that's got one of these outliers.
* Very simple point: the left corner taker will get a false bump Vs the right corner taker.
* Will want to plot xT maps with and without this bit of analysis
* And the lovely thing is: we can just bake this in as part of our prior -> synthetic counts.

```python
pitch = VerticalPitch(pitch_type='opta', line_zorder=2, pitch_color='#f4edf0', half=True)

# refreshing the shots dataframe and adding a 1/0 binary goal flag
df_shots = df_opta.loc[df_opta['eventType'] == 'shot']
df_shots['goal'] = df_shots.eventSubType.apply(lambda x: 1 if x == 'Goal' else 0)

# fontsize
fs = 15

# setting up the figure
fig, ax = pitch.draw(figsize=(20, 30))
fig.set_facecolor('#f4edf0')

# calculating a mean bin statistic using the same method as above but using a different statistic
bin_statistic_mean = pitch.bin_statistic(df_shots.x1, df_shots.y1, values=df_shots.goal, statistic='mean', bins=b)
bin_statistic_mean['statistic'] = np.nan_to_num(bin_statistic_mean['statistic'])
bin_statistic_mean['statistic'] = (bin_statistic_mean['statistic'] * 100).astype(np.int32)

# plotting the mean counts
pitch.heatmap(bin_statistic_mean, ax=ax, cmap=cm.coolwarm, edgecolor='#22312b')

# adding the labels
labels = pitch.label_heatmap(bin_statistic_mean, color='#f4edf0', fontsize=fs, ax=ax, ha='center', va='center', zorder=3)

### outliers: over 60% of shots being scored and under 20 shots being taken from that cell
## using numpy's 2d binned statistic function as we need to expand the binnumbers to trace back to the original data to plot the shots (goals) in the outlier bins
bs_mean = binned_statistic_2d(df_shots.x1, df_shots.y1, values=df_shots.goal, statistic='mean', bins=b, expand_binnumbers=True, range=[[0,100],[0,100]])
bs_counts = binned_statistic_2d(df_shots.x1, df_shots.y1, values=df_shots.goal, statistic='count', bins=b, expand_binnumbers=True, range=[[0,100],[0,100]])

# extracting the outlier indices using the nonzero() method - very handy!
x_outliers, y_outliers = ((bs_mean.statistic >= 0.1) & (bs_counts.statistic <= 30)).nonzero()
#x_outliers, y_outliers = (bs_counts.statistic > 300).nonzero()

# producing outlier dataframe
df_outliers = pd.DataFrame(np.array([x_outliers+1,y_outliers+1]).T, columns=['bin_x_outlier','bin_y_outlier'])
df_outliers['outlier'] = True

# using the key extra bit of data that's in numpy's 2d_binned_statistic method that's not in the mpl soccer method: the raw bins
# attributing these bins to the original dataframe
# this is critical so that we can identify the shots in the original data that are being identified as outliers so that we can plot them
df_shots['bin_x'] = bs_mean.binnumber[0]
df_shots['bin_y'] = bs_mean.binnumber[1]

# joining the outliers onto the original shots dataframe
df_shots = df_shots.merge(df_outliers, left_on=['bin_x', 'bin_y'], right_on=['bin_x_outlier','bin_y_outlier'], how='left')

#Â making another outlier dataframe, but this time we're querying the shots data frame for outlier goals, and we'll use this subset of shot/goals to plot and visualise the outliers
df_outliers_ = df_shots.loc[(df_shots['outlier'] == True) & (df_shots['eventSubType'] == 'Goal')]

# and getting rid of the shots directly on the goal line for the plot of the outliers (as they look messy)
# creating a square region that approximately covers the 6 yard box
y_start = 37
y_end = 100-y_start
x_start = 90
x_end = 100

# applying extraction region
df_outliers_ = df_outliers_.loc[~((df_outliers_['x1'] > x_start) & (df_outliers_['y1'] > y_start) & (df_outliers_['x1'] < x_end) & (df_outliers_['y1'] < y_end))]

# plotting outliers
pitch.scatter(df_outliers_.x1, df_outliers_.y1, ax=ax, label='outliers', alpha=0.65, marker='football', c='white')


# recolouring the text of the middle bins to black so they stand out a little bit more
mask = (bin_statistic_mean['statistic'] < 70) & (bin_statistic_mean['statistic'] >= 9)
bin_statistic_mean['statistic'] = bin_statistic_mean['statistic'].astype(str)
bin_statistic_mean['statistic'][~mask] = ''
annotation = pitch.label_heatmap(bin_statistic_mean, color='black', fontsize=fs, ax=ax, ha='center', va='center', zorder=4)

# saving figure
fig.savefig('/Users/christian/Desktop/University/Birkbeck MSc Applied Statistics/Project/Plots/Synthetic Shots/Real Shot Success - All Opta.png', bbox_inches = 'tight', pad_inches = 0.2)
```

## Generating Random Data

* We're going to randomly sprinkle shots in the grid
* Then we'll get rid of anything in the box, and one box outside of the box
* Then we'll apply a Binomial random sample given a success probability to those shots to give us a goal flag

```python
# setting random seed
np.random.seed(9)

# pitch dimensions
pl, pw = 105, 68
# number of initial fake shots
# this number just needs to be sufficiently high for now
# we're going to filter away lots of this number for fake shots in the goal area - so at the end we will pick a number of shots equal to the number of shots in the original dataset
# and this will give our weight of prior
N = 100000

x = np.arange(0, 100, 0.1)
y = np.arange(0, 100, 0.1)
# bootstrapped samples from x and y (sampling with replacement)
x_boot = np.random.choice(x, size=N, replace=True)
y_boot = np.random.choice(y, size=N, replace=True)
```

### Initialisation of synthetic shots

```python
# need to be careful with the range here to ensure all of the binning matches up
#Â we'll bootstrap over the opta co-ord range [0,100]x[0,100] and then transform to x1_m and y1_m afterwards, so that we have identical ranges
# with the analysis above
synthetic_shots = binned_statistic_2d(x_boot, y_boot, values=x_boot, statistic='count', bins=b, expand_binnumbers=True, range=[[0,100],[0,100]])

df_synthetic = pd.DataFrame({'x1': x_boot, 'y1': y_boot})
df_synthetic['bin_x'] = synthetic_shots.binnumber[0]
df_synthetic['bin_y'] = synthetic_shots.binnumber[1]
df_synthetic['x1_m'] = pl * (df_synthetic.x1 / 100)
df_synthetic['y1_m'] = pw * (df_synthetic.y1 / 100)
```

### Synthetic goals: Setting background prior of 1% of these shots will be scored

> Feels like a very sensible prior: to say that 1 in 100 will go in.

```python
p_goal = 0.004

df_synthetic['goal'] = np.random.binomial(1, p_goal, N)

df_synthetic.goal.mean()
```

### Removing synthetic shots from bins where they're not needed

#### 1. Removing them where there had been more than 50 shots

```python
# setting threshold for each bin where anything above, and we don't need to apply a prior
shot_threshold = 50

# getting the indices of the bins where we will be applying prior
x_threshold, y_threshold = (bs_counts.statistic > shot_threshold).nonzero()

# transforming this into a dataframe
df_shot_threshold = pd.DataFrame(np.array([x_threshold+1,y_threshold+1]).T, columns=['bin_x','bin_y'])
df_shot_threshold['remove'] = True

# and now joining this onto our synthetic shots (to remove these bins)
df_synthetic = df_synthetic.merge(df_shot_threshold, how='left', on=['bin_x','bin_y'])

# and getting rid of the rows to remove
df_synthetic = df_synthetic.loc[pd.isna(df_synthetic['remove']) == True].drop(columns=['remove'])
```

#### 2. And now removing fake shots from central area

```python
df_synthetic
```

```python
# creating a square region that's central to the goal mouth and extends slightly beyond the box
y_start = 37
y_end = 100-y_start
x_start = 80
x_end = 100

df_synthetic = df_synthetic.loc[~((df_synthetic['x1'] > x_start) & (df_synthetic['y1'] > y_start) & (df_synthetic['x1'] < x_end) & (df_synthetic['y1'] < y_end))]
```

#### 3. And removing goals from own-half shots

```python
df_synthetic.loc[df_synthetic['x1'] <= 60, 'goal'] = 0
```

#### 4. And finally weighting our prior

* We're using a Beta-Binomial conjugate analysis
* It feels appropriate to say that we will weight this prior equally with the data, because we're weighting our data with domain expertise, which encapsulates the knowledge of the shots the players **don't take** because they know it's broadly pointless to shoot from there.
* So we will take a random sample, **without replacement**, from the synthetic data that's the same size as the number of shots in our dataset

```python
df_synthetic_prior = df_synthetic.sample(n=len(df_shots), replace=False)

len(df_synthetic_prior)
```

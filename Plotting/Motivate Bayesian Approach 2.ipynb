```r
library(repr)
```

```r
df <- read.csv('grouped_shot_distances.csv')
```

```r
head(df)
```

```r
lm.c1_m <- lm(c1_m ~ GN, data=df)
summary(lm.c1_m)

c1_m.pred <- predict.lm(lm.c1_m, data.frame(GN=df$GN))

#plot(df$GN, df$c1_m)
#lines(df$GN, c1_m.pred, lwd=5, col='blue')
```

```r
lm.x_dist_goal <- lm(x_dist_goal ~ GN, data=df)
summary(lm.x_dist_goal)

x_dist_goal.pred <- predict.lm(lm.x_dist_goal, data.frame(GN=df$GN))

#plot(df$GN, df$x_dist_goal)
#lines(df$GN, x_dist_goal.pred, lwd=5, col='red')
```

### Making a nice plot for the report

```r
#png(file="/Users/christian/Desktop/University/Birkbeck MSc Applied Statistics/Project/Plots/Bayesian Motivation/Decreasing Distances To Goal.png", width=6, height=12, units="in", res = 300)

options(repr.plot.width=6, repr.plot.height=12)

# making a 2x1 plot
par(mfrow = c(2, 1))

# central distance
plot(df$GN, df$c1_m, xlab = 'Season', ylab = 'y Distance from Centre of Pitch (m)', xaxt='n')
lines(df$GN, c1_m.pred, lwd=5, col='blue')
abline(v = seq(0,176,44), col='gray')
axis(side=1, tick = seq(0,176,44), labels = c('2017/18','2018/19','2019/20','2020/21'), at = seq(22,154,44))

# distance to goal
plot(df$GN, df$x_dist_goal, xlab = 'Season', ylab = 'x Distance to Goal (m)', xaxt='n')
lines(df$GN, x_dist_goal.pred, lwd=5, col='red')
abline(v = seq(0,176,44), col='gray')
axis(side=1, tick = seq(0,176,44), labels = c('2017/18','2018/19','2019/20','2020/21'), at = seq(22,154,44))

#dev.off()
```

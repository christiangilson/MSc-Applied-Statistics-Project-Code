```python
import pandas as pd
import numpy as np
import os

repo = r'../Data/Opta/EPL Data/Events'
```

```python
os.listdir(repo)
```

```python
df_1718 = pd.read_csv(os.path.join(repo, 'df_subevents_EPL_1718.csv'))
df_1819 = pd.read_csv(os.path.join(repo, 'df_subevents_EPL_1819.csv'))
df_1920 = pd.read_csv(os.path.join(repo, 'df_subevents_EPL_1920.csv'))

df_1718['season'] = '17/18'
df_1819['season'] = '18/19'
df_1920['season'] = '19/20'
```

```python
len(df_1718), len(df_1819), len(df_1920)
```

```python
df_1718.columns
```

```python
df_1718.groupby(['eventType','eventSubType'])\
        .agg({'matchId':'count'})\
        .rename(columns={'matchId':'numEvents'})\
        .reset_index()\
        .sort_values(['eventType','numEvents'], ascending=[True, False])
```

```python
df = pd.concat([df_1718, df_1819, df_1920])
```

```python
df_summary = df.groupby('season').agg({'matchId':'nunique', 'playerId':'count'}).rename(columns={'matchId':'numMatches','playerId':'numEvents'}).reset_index()
```

```python
df_summary['eventsPerMatch'] = df_summary['numEvents'] / df_summary['numMatches']

df_summary
```

```r
library(ggplot2)
library(repr)

options(repr.plot.width=12, repr.plot.height=12)
```

```r
df <- read.csv('Kane.csv')
df['ConcentrationRank'] <- factor(df$seasonRank)
df
```

```r
ggplot(df, aes(y=xT_Pc, x=season)) + 
    geom_bar(aes(fill=ConcentrationRank), position="dodge2", stat="identity") +
    geom_text(aes(y=xT_Pc/2, fill=ConcentrationRank, label=otherName), angle=90, color='white', position = position_dodge(width = 0.9)) +
    labs(x='Season', y = 'Harry Kane Paired Concentration of Spurs xT (%)') +
    theme_light(base_size = 20) +
    guides(fill=guide_legend(title="Kane Concentration Rank")) +
    theme(legend.position="bottom")

ggsave('/Users/christian/Desktop/University/Birkbeck MSc Applied Statistics/Project/Plots/Paired xT/KaneConcentration.png', dpi=300, width = 10, height=10, units='in')
```

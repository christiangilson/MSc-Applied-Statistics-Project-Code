```r
library(ggplot2)
library(repr)

options(repr.plot.width=12, repr.plot.height=12)
```

```r
df = read.csv('Opta_V_Wyscout_DQ_for_R.csv')
head(df)
```

```r
v <- df[df['Ratio'] == 'Minutes Played','value']

mean(v[is.finite(v)])
```

```r
df$Ratio <- factor(df$Ratio, levels = c("Dribble", "Pass", "Shot", "Minutes Played"))

# Violin plot
p <- ggplot(df, aes(x=Ratio, y=value, color=Ratio, fill=Ratio)) +
    geom_violin(trim=TRUE) +
    labs(x='Event Type', y = 'Opta / Wyscout Ratio') +
    geom_jitter(shape=16, position=position_jitter(0.25), alpha=0.2) + 
    stat_summary(fun=mean, geom="point", size=2, col='yellow') +
    geom_hline(yintercept=1, linetype="dashed", color = "black", size=0.5, alpha=0.5) + 
    theme_light(base_size = 20) +
    theme(legend.position = "none")

p

ggsave('/Users/christian/Desktop/University/Birkbeck MSc Applied Statistics/Project/Plots/DQ/ViolinDQRatios.png', dpi=300, width = 10, height=10, units='in')
```

```r
df_opta <- read.csv('R_Opta_Mins.csv')
df_wyscout <- read.csv('R_Wyscout_Mins.csv')
```

```r
o <- ggplot(df_opta, aes(x=minsPlayed)) +
    geom_histogram(binwidth=3, fill="#02adef", color="#e9ecef", alpha=0.9) +
    labs(x='Opta Minutes Played', y = 'Frequency') +
    theme_light(base_size = 15) +
    scale_x_continuous(limits=c(-3, 106))
o
```

```r
w <- ggplot(df_wyscout, aes(x=minutesPlayed)) +
    geom_histogram( binwidth=3, fill="black", color="#e9ecef", alpha=0.9) +
    labs(x='Wyscout Minutes Played', y = 'Frequency') +
    theme_light(base_size = 15) +
    scale_x_continuous(limits=c(-3, 106))
w
```
